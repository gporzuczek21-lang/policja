```python
#2
d1=df.pivot_table(values='id',index='race',columns='signs_of_mental_illness',aggfunc="count")
d1
```

```python
d1['mental_illness_ratio'] = d1.apply(lambda row: row[True] / (row[True] + row[False]), axis=1)
d1
```

```python
max = d1['mental_illness_ratio'].idxmax()
d1.loc[max]
```
