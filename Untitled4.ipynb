```python
import pandas as pd

populacja = pd.read_csv("populacja_stanow.csv")
skroty = pd.read_csv("skroty_stanow.csv")
dane = pd.read_csv("police.csv")
df = pd.merge(populacja, skroty, on='state', how='inner')
dane = dane.rename(columns={'state': 'abbreviation'})
dane1=dane['abbreviation'].value_counts()
df1 = df.merge(dane1, on='abbreviation', how='left')
df1['incidents_per_1000'] = df1['count'] / df1['population'] * 1000
df1
```
